```python
sigma = xre.corr()
```

```python
from numpy.linalg import eig
_lambda,eig_vector=eig(np.array(sigma))
_frame = pd.DataFrame({"values":_lambda,"vector":[str(i) for i in np.transpose(eig_vector)]})
sigma
```

```python
_frame = _frame.sort_values('values',ascending = False)
_frame
```

```python
PCA_data = pd.DataFrame()
for i in range(len(_lambda)):
  PCA_data['PCA '+str(i+1)] = np.dot(np.transpose(eig_vector)[i],np.transpose(xre))
variances = [np.var(PCA_data[i]) for i in PCA_data.columns]
np.around(np.array(variances),6)
variances.sort(reverse = True)
```

```python
PCA_data.shape
```

```python
plt.figure(figsize = (10,6))
plt.bar(range(1,len(variances)+1),sorted(variances,reverse = True),edgecolor='black')
plt.plot(range(1,len(variances)+1),sorted(variances,reverse = True),"*-",color = 'red')
plt.title('SCREE PLOT OF PRINCIPAL COMPONENT VARIANCE')
plt.xlabel("components")
plt.ylabel("variance")
```

```python
from scipy.stats import pearsonr
pca_dict = dict()
for pcas in PCA_data:
  pca_dict[pcas] = []
  for col in xre.columns:
    corr =pearsonr(PCA_data[pcas],xre[col])[0]
    pca_dict[pcas].append(corr)
```

```python
interpret = pd.DataFrame(pca_dict,index = xre.columns)
interpret[['PCA 1','PCA 2','PCA 3','PCA 4','PCA 5','PCA 6']]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=15)
pca_result =pd.DataFrame( pca.fit_transform(xre))
f=pd.concat([pca_result,yre],axis=1)
f1=f.loc[f['Hazardous']==0]
f1.reset_index(inplace=True)

color = ['yellow']
def biplot(score, coeff, labels=None, scaling_factor=1.0):
    plt.figure(figsize=(8, 8))
    coeff = np.transpose(coeff)
    plt.scatter(score[0], score[1], alpha=0.7)
    for i in range(coeff.shape[1]):
        plt.arrow(0, 0, coeff[0, i] * scaling_factor, coeff[1, i] * scaling_factor,
                  color='r', alpha=0.5, head_width=0.5, head_length=0.1,width = 0.05)
        if labels is not None:
            plt.text(coeff[0, i] * scaling_factor * 1.2, coeff[1, i] * scaling_factor * 1.2,
                     labels[i], color='k', ha='center', va='center',fontsize = 10)

    plt.xlabel(f'Principal Component 1 ')
    plt.ylabel(f'Principal Component 2 ')
    plt.grid()

# Create an extended biplot with scaling factor 2.0
biplot(f1, np.transpose(pca.components_), labels=xre.columns, scaling_factor=25.0)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
model2=nb.fit(x_train,y_train)
pred=nb.predict(x_test)
c2=confusion_matrix(y_test,pred)
d=ConfusionMatrixDisplay(c2,display_labels=[0,1])
d.plot()
plt.show()
print(c2)
print(classification_report(y_test,pred))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
clf = [('dc',dc),('rfc',rfc),('knn',knn),('nb',nb)]
lr=LogisticRegression()
stack_model = StackingClassifier( estimators = clf,final_estimator = lr)
model=stack_model.fit(x_train,y_train)
pred=stack_model.predict(x_test)
c4=confusion_matrix(y_test,pred)
d=ConfusionMatrixDisplay(c4,display_labels=[0,1])
d.plot()
plt.show()
print(c4)
print(classification_report(y_test,pred))
```
